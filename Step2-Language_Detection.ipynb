```python
def language_id(text):
    # tokenize
    tokens = text.split(' ')
    new_tokens = []
    for token in tokens:
        if token[-1] in punctuation:
            new_tokens.append(token[:-1])
            new_tokens.append(token[-1])
        else:
            new_tokens.append(token)   
    # language tag
    lang_tags = []
    for token in new_tokens:
    try:
        token.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        lang_tags.append("Hin")
    else:
       lang_tags.append("Eng")
    
    
    return
```
