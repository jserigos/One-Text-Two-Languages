# Language Identification

## By character encoding


```python
import pandas as pd
```

```python
with open('Data/Significance_of_festivals_in_India.txt', 'r',encoding="utf-8") as myfile:
    data=myfile.read().replace('\n', '')
```

```python
tokens = data.split(' ')
```

```python
tokens
```

```python
lang_tags = []
#i = 0
tokens = tokens[1:]
for token in tokens:
    try:
        token.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        lang_tags.append("Hin")
    else:
       lang_tags.append("Eng")
```

```python
def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True
isEnglish(tokens[39])
```

```python
for x, y in zip(tokens, lang_tags):
    print(x, y)
for i, y in enumerate(tokens):
    print(i, y)
```

```python
df = pd.DataFrame(
    {'word': tokens,
     'language': lang_tags,
    })
```
