# Unicode Files
Checking to see how easy it is to process utf-8 formatted Hindi English 

```python
# Lets import modules needed
import sys
import math
import numpy as np
import pandas as pd
from itertools import groupby
from collections import Counter
import csv
```

```python
with open('Significance_of_festivals_in_India.txt', 'r',encoding="utf-8") as myfile:
    data=myfile.read().replace('\n', '')
```

```python
data = data.replace('(','')
data = data.replace(')','')
```

```python
tokens = data.split(' ')
```

```python
tokens
```

```python
lang_tags = []
i = 0
tokens = tokens[1:]
for token in tokens:
    t = ord(token[0])
    if (((t > 0x0041) and(t < 0x005A)) or ((t > 0x0061) and(t < 0x006A))):
       lang_tags.append("Eng")
    else:
       lang_tags.append("Hin")
```

```python
lang_tags
```

```python
df = pd.DataFrame(
    {'word': tokens,
     'language': lang_tags,
    })
```
